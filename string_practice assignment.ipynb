**Question - 4:**
Given a string, count how many times the letter 'a' appears in it.

Example:

text = "data analysis"


Expected Output: 3

***Hint:***
Use loop or count().

```python
#write your code here 
text = "data analysis"
count = 0
for i in text:
    if i == "a":
        count += 1
print(count)    
```

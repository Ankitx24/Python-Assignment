## 1. Calculate the Total Bill
A restaurant bill is ₹450. GST is 18%. Calculate the final bill amount.

```python
# Write your solution here
bill = 450
gst = 0.18 * bill

print(gst)

Final_bill = bill + gst
print(Final_bill)
```

Grocery Bill

```python
Veg = 345
v_gst = 0.08 * Veg

Milk = 60
m_gst = 0.10*Milk

Sweet = 390
s_gst = 0.12*Sweet

final_bill = Veg + v_gst + Milk + m_gst + Sweet + s_gst

Discount = final_bill - (0.10 * final_bill)

print(final_bill)
print(Discount)
```

## 2. Combine Two Numbers
Create two variables `a = 7` and `b = 3` and compute sum, difference, product, and a^b.

```python
# Write your solution here
a = 7
b = 3

Sum = a+b
diff = a-b
Prod = a*b
power = a^b

print(Sum)
print(diff)
print(Prod)
print(power)
```

## 3. Temperature Converter

## 4. Movie Ticket Cost Calculator
Movie ticket = 180, popcorn = 120. Calculate total for 2 tickets + 1 popcorn and 3 tickets + 0 popcorn.

```python
# Write your solution here
Movie_ticket_price = 180
Popcorn_price = 120

bill_one = 2*Movie_ticket_price + 1*Popcorn_price

bill_two = 3*Movie_ticket_price + 0*Popcorn_price

print(bill_one)
print(bill_two)

Total = bill_one + bill_two

print(f"Total_Cost :{Total}")
```

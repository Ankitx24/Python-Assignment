## Problem 5: Movie Ticket Discount

**Question:** Apply discount based on age.

**Approach:** Use ordered elif statements from smallest to largest age group.

Apply discounts based on age:
- Age < 12 → 50% discount
- Age between 12 and 18 → 30% discount
- Age between 18 and 60 → no discount
- Age > 60 → 40% discount

```python
# Enter your answer here
Age = int(input("Enter your Age :"))
if Age < 12:
    print(f"{Age} years you got 50% discount")
elif 12 <= Age < 18:
    print(f"{Age} years you got 30% discount")
elif 18 <= Age < 60:
    print(f"{Age} years you got No Discount") 
elif Age > 60:
    print(f"{Age} years you got 40% discount")       
```

## Problem 7: Delivery Charge System

**Question:** Calculate final delivery charge based on prime membership, amount, and distance.

**Approach:** Start with prime-check, then inside it apply nested if blocks.

Inputs:
- order_amount (int)
- is_prime_member (yes/no)
- distance_km (float)

Rules:
- If prime member → Free delivery
- Else:
  - If amount ≥ 500:
    - If distance ≤ 5 km → Free
    - Else → ₹20 extra
  - Else:
    - If distance ≤ 5 km → ₹30
    - Else → ₹50

```python
# Enter your answer here
order_amount = int(input("Eneter Total amount:"))
prime_member = input("Do u have premium :")
distance = int(input("Enter distance in Km :"))

if prime_member == 'yes':
    print("Free Delivery for prime member")
if order_amount >= 500:
    if distance <= 5:
        print("Total Amount",order_amount) 
    else:
        print("Total + Rupee 20 extra =",order_amount + 20)     
else:
    if distance <= 5:
        print("Total + Rupee 30 extra =",order_amount + 30)
    else:
        print("Total + Rupee 50 extra =",order_amount + 50)    
```

## Problem 8: ATM Withdrawal Simulation

**Question:** Simulate ATM rules: multiples of 100, sufficient balance, manager approval.

**Approach:** Check errors first, then nested conditions for amount > 10000.

Inputs:
- balance
- amount

Logic:
- If amount % 100 != 0 → reject (ATM works with ₹100 multiples)
- If amount > balance → reject
- Else:
  - if amount > 10000 → ask for manager approval
  - else → withdraw success

```python
# Enter your answer here
balance = int(input("Enter Total balance in your Bank:"))
amount = int(input("Enter amount to withdraw :")) 

if amount % 100!= 0:
    print("Enter valid amount it must be multiple of 100")
elif amount > balance:
    print("Insufficient Balance") 
else:
    if amount > 10000:
        print("Ask for manager Approval")
    else:
        print("Withdraw Successful",amount)           

```

## Problem 9: Employee Performance Rating

**Question:** Assign performance rating based on sales, feedback, attendance.

**Approach:** Prioritize attendance rule, then apply nested conditions.

Inputs:
- sales (₹)
- feedback (0-5 scale)
- attendance (%)

Logic:
- If attendance < 70 → 'Poor'
- Else:
  - If sales > 100000 AND feedback > 4.5 → 'Excellent'
  - If sales > 70000 OR feedback > 4 → 'Good'
  - Else → 'Average'

```python
# Enter your answer here
sales = int(input("Enter Total Sales in Rupee:"))
feedback = float(input("Enter Feedback :"))
attendence = int(input("Enter Total Attendence in % :"))

if feedback < 0 or feedback > 5:
    print("Invalid feedback rating! Please enter a value between 0 and 5.")
elif attendence < 70:
    print("Poor")
else:
    if sales > 100000 and 5 >= feedback >= 4.5:
        print("Excellent")
    if sales > 70000 or 4.5 > feedback >= 4:
        print("Good")
    else:
        print("Average")      
```

```python
#Calculator

num1 = float(input("Enter first number:"))
num2 = float(input("Enter second number:"))

operator = input("Enter any operator like +,-,*,/:")

if operator == "+":
  print(num1 + num2)
elif operator == "/" and num2 == 0:
  print("Infinity")  
elif operator == "-":
  print(num1 - num2) 
elif operator == "*":
  print(num1 * num2)  
elif operator == "/":
  print(num1 / num2)
else:
  print("Enter a valid choice like +,-,*,/")
```

```python
#Stone Paper Scissor

import random

choices = ["stone","paper","scissor"]

user = input("Enter your choice (stone/paper/scissor) : ")
computer = random.choice(choices)

print(f"Computetr chooses : {computer}")

if user == computer:
    print("Its a draw..")
elif (user == "stone" and computer == "scissor") or (user == "paper" and computer == "stone") or \
    (user == "scissor" and computer == "paper"):
    print("you win ...")
else:
    print("Computer wins...")  
```

```python
# Coin Toss Guess

import random

choices = ["head","tail"]

user = input("What's your call (HEADS/TAIL) :").lower()
computer = random.choice(choices)

print(f"Computer Calls : {computer}")

if user == computer:
    print("You WIN the Toss....")
else:
    print("You Loss Better Luck Next Time...")    
```

```python
# Dice Roll Challenge with highest numbers

import random

print("Dice Game - 3 Rolls Each : ")

# random.randint(a, b) means returns a random integer between a and b (including BOTH a and b).
user_rolls = [random.randint(1,6) for i in range(6)]          
computer_rolls = [random.randint(1,6) for i in range(6)]

print("My Rolls : ",user_rolls)
print("Computer Rolls : ",computer_rolls)

my_total = sum(user_rolls)
computer_total = sum(computer_rolls)

print("My total = ",my_total)
print("computer total = ",computer_total)

if my_total > computer_total:
    print("I Won The Challenge...")
elif my_total < computer_total:
    print("Computer Wins The Challenge..")
else:
    print("Its Draw Roll Once Again....")        
```

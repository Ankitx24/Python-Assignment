## Question 16 (Hard)

**Title:** Check Prime Number

Take a number as input and check whether it is prime or not.

### Approach:
Loop from 2 to N-1 and use `break` if divisible.


```python
# write your code here
n = int(input("Enter a number ="))
for i in range(2,n-1):
    if n % i == 0:
            print(f"{n} is not Prime Number")
            break
else:
    print(f"{n} is prime Number")    
```

## Question 17 (Hard)

**Title:** Prime Numbers Between 1 and N

### Problem Statement:
Take N as input and print all prime numbers between 1 and N.

### Approach:
Use nested loops and a flag variable for prime checking.


```python
# write your code here
start = int(input("Enter start: "))
end = int(input("Enter end: "))

print("Prime numbers:", end=" ")

for num in range(start, end + 1):
    if num > 1:
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                break
        else:
            print(num, end=" ")
```

## Question 20 (Hard)

**Title:** Perfect Number Check

### Problem Statement:
Take a number as input and check whether it is a perfect number or not.

### Approach:
Find sum of all factors using a loop and compare with the number.


```python
# write your code here
n = int(input("Enter a number: "))

div_sum = 0
for i in range(1, n//2 + 1):
    if n % i == 0:
        div_sum += i

if div_sum == n:
    print(f"{n} is a Perfect Number")
else:
    print(f"{n} is NOT a Perfect Number")
```

## 1. What Are Conditionals?
Conditionals help Python make decisions based on conditions.

## 2. Basic `if` Example

```python
age = 20
if age >= 18:
    print("Adult")
```

## 3. `if` and `else` Example

```python
age = 15
if age >= 18:
    print("Adult")
else:
    print("Minor")
```

## 4. `elif` Example

```python
marks = 72
if marks >= 90:
    print("A Grade")
elif marks >= 75:
    print("B Grade")
else:
    print("C Grade")
```

## 5. Comparison Operators

| Operator | Meaning |
|---------|---------|
| == | equal to |
| != | not equal to |
| > | greater than |
| < | less than |
| >= | greater than or equal |
| <= | less than or equal |

## 6. Simple Examples

```python
num = 6
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
temp = 40
if temp > 35:
    print("It's hot")
else:
    print("Normal weather")
```
